```python
from bokeh.plotting import figure
from bokeh.io import output_notebook, push_notebook, show
from bokeh.transform import cumsum
from math import pi
```

### Pie chart of Stores that are open on Saturday or Not

```python
data = merged_data["IsOpen.Sat"].value_counts().reset_index(name='value').rename(columns={'index':'boolean'})
data['angle'] = data['value']/data['value'].sum() * 2*pi
data['percentage'] = ["{:.0f}%".format(d) for d in data['value']/data['value'].sum() * 100]
data['color'] = ["Turquoise", "Silver"]
data
```

```python
output_notebook()
```

```python
f = figure(plot_height=350, title="Pie chart of Stores that are open on Saturday or Not", toolbar_location=None,
           tools="hover", tooltips="@boolean: @percentage", x_range=(-0.5, 1.0))

f.wedge(x=0, y=1, radius=0.4, start_angle=cumsum('angle', include_zero=True), end_angle=cumsum('angle'),
        line_color="white", fill_color='color', legend='boolean', source=data)


f.axis.axis_label=None
f.axis.visible=False
f.grid.grid_line_color = None
```

### Please hover on the chart to see the percentage for each labels

```python
handle = show(f, notebook_handle=True)
push_notebook(handle=handle)
```

### Pie chart of Stores that are open on Sunday or Not

```python
s_data = merged_data["IsOpen.Sun"].value_counts().reset_index(name='values').rename(columns={'index':'booleans'})
s_data['angles'] = s_data['values']/s_data['values'].sum() * 2*pi
s_data['percentages'] = ["{:.0f}%".format(d) for d in s_data['values']/s_data['values'].sum() * 100]
s_data['colors'] = ["Turquoise", "Silver"]
s_data
```

```python
p = figure(plot_height=350, title="Pie chart of Stores that are open on Sunday or Not", toolbar_location=None,
           tools="hover", tooltips="@booleans: @percentages", x_range=(-0.5, 1.0))

p.wedge(x=0, y=1, radius=0.4, start_angle=cumsum('angles', include_zero=True), end_angle=cumsum('angles'),
        line_color="white", fill_color='colors', legend='booleans', source=s_data)

p.axis.axis_label=None
p.axis.visible=False
p.grid.grid_line_color = None
```

### Please hover on the chart to see the percentage for each labels

```python
handles = show(p, notebook_handle=True)
push_notebook(handle=handles)
```
