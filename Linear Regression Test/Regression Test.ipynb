```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("weight-height.csv", header=None, skiprows=1)
new_df = df.drop(columns=0)
new_df = df[[2, 1]].astype(float)


X = new_df[ 2]
Y = new_df[ 1]


#plot to see what it looks like
plt.scatter(X, Y)
plt.show()

# soling for the value of a and b

denominator = X.dot(X) - X.mean() * X.sum()
a = (X.dot(Y) - Y.mean() * X.sum() ) / denominator
b = (Y.mean() * X.dot(X) - X.mean() * X.dot(Y) ) / denominator

#to calculte the predicted y
Yhat = a * X + b

#plot it all
plt.scatter(X, Y)
plt.plot(X, Yhat)
plt.show()

#r-squared value calculation
d1 = Y - Yhat
d2 = Y - Y.mean() 
r2_one = 1 - d1.dot(d1) / d2.dot(d2)
print("the r-squared value for quastion 1 is: ", r2_one)
```
