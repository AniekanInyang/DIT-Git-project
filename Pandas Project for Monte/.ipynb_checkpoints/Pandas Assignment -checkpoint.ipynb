```python
#8. Create a list of stores that close by 11pm
```

```python
merged_version.rename(columns={"ThruTime.MF":"ThruTime_MF", "Name_x":"Store_Name", "Name_y":"Store_Name", "LocationMilestones.LastRemodelDate":"Remodel_Date", "Address.City_x":"Address_City","Address.City_y":"Address_City"}, inplace=True)
```

```python
closing_time = merged_version["ThruTime_MF"] == "23:00:00"
closing_time.head()
```

```python
booleans = []
for result in closing_time:
    if result== True:
        booleans.append(True)
    else:
        booleans.append(False)
booleans[0:]
    
```

```python
filtered = pd.Series(booleans)
Shops_that_close_by_11pm = merged_version[filtered]
```

```python
Shops_that_close_by_11pm.head()
```

```python
List_of_shops_that_close_by_11pm = []
```

```python
for rows in Shops_that_close_by_11pm.itertuples():
    the_list = [rows.Store_Name]
    
    List_of_shops_that_close_by_11pm.append(the_list)
    
List_of_shops_that_close_by_11pm
```

```python
#9. Get a list of stores that were remodeled and their remodeled date. Save these in a dictionary 
```

```python
Remodel_Dates = merged_version["Remodel_Date"]
```

```python
Remodeled = Remodel_Dates>= "2000"
Remodeled.head()
```

```python
Remodelling_Dates = []
for result in Remodeled:
    if result == True:
        Remodelling_Dates.append(True)
    else:
        Remodelling_Dates.append(False)
Remodelling_Dates
```

```python
filtered1 = pd.Series(Remodelling_Dates)
Stores_that_were_remodeled = merged_version[filtered1]
Stores_that_were_remodeled
```

```python
List_of_shops_that_were_remodeled_saved_in_disctionary = []
```

```python
for rows in Stores_that_were_remodeled.itertuples():
    the_list = [rows.Store_Name, rows.Remodel_Date]
    
    List_of_shops_that_were_remodeled_saved_in_disctionary.append(the_list)
    
List_of_shops_that_were_remodeled_saved_in_disctionary    
```

```python
dict(List_of_shops_that_were_remodeled_saved_in_disctionary)
```
