# Preprocessing for LDA
Task: Use Count Vectorization to create a vectorized document sparse matrix. Use a max_df=0.9 and min_df=2 parameters.

```python
# Import the required packages i.e Countvectorizer, LDA

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

```python
# Using the CV object, Fit and transform the countvectorizer 
# on the articles using the params giving above
cv = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
dtm = cv.fit_transform(art['articles'])
dtm
```

```python
#Fit LDA object to the sparse matrix fron the count vectorizer
# Extracting 10 TOPICS with a RANDOM STATE of 10
from sklearn.decomposition import LatentDirichletAllocation
LDA = LatentDirichletAllocation(n_components=10,random_state=10)
LDA.fit(dtm)
```

```python
# View the the TOP 20 words of the all Topics 
for index,topic in enumerate(LDA.components_):
    print(f'THE TOP 15 WORDS FOR TOPIC #{index}')
    print([cv.get_feature_names()[i] for i in topic.argsort()[-20:]])
    print('\n')

```

```python
# Transform the sparse data on the fitted LDA object
topic_results = LDA.transform(dtm)

# Create a new column called 'Topic_lda' and save the Topic id's in it

art["Topic_lda"] = topic_results.argmax(axis=1)
```

```python
# View the dataframe now
art.head()
```

```python
# Create a label dictionary for each topic, kinda like mine,
#please go through the words in each topic to figure out a fitting 
#label for that topic and create a dictionary mapping each topic id 
#to its corresponding label (which is at your discretion)

label_dict = {
    0:'security',
    1:'economy',
    2:'economy',
    3:'election',
    4:'economy',
    5:'football',
    6:'election',
    7:'stock',
    8:'social',
    9:'football'
}

# there seem to be just 6 unique labels
```

```python
# Create a new column called 'lda_labels' based on the map created above
art['lda_Label'] = art['Topic_lda'].map(label_dict)
art.head()
```

# Optional Challenge: Text Generation with Neural Networks

```python
# Read in the oliver-twist-0-9.txt text
with open('oliver-twist-0-9.txt') as c:
    fields = c.read()

    
# Run fields to see the contents of contacts.txt:
fields
```

### Tokenize and Clean Text

```python
# Import standard libraries
import spacy
nlp = spacy.load('en',disable=['parser', 'tagger','ner'])
# Increase the allowable max lentgh
nlp.max_length = 1198623
```

```python
# Run this function to seperate punctuations
def separate_punc(doc_text):
    
    return [token.text.lower() for token in nlp(doc_text) if token.text not in '\n\n \n\n\n!"-#$%&()--.*+,-/:;<=>?@[\\]^_`{|}~\t\n ']
tokens = separate_punc(fields)
```

### Create Sequences of Tokens

```python
# organize into sequences of tokens
train_len = 25+1 # 50 training words , then one target word

# Empty list of sequences
text_sequences = []

for i in range(train_len, len(tokens)):
    
    # Grab train_len# amount of characters
    seq = tokens[i-train_len:i]
    
    # Add to list of sequences
    text_sequences.append(seq)
```

### Keras Tokenization

```python
from keras.preprocessing.text import Tokenizer
import numpy as np
```

```python
# integer encode sequences of words
tokenizer = Tokenizer()
tokenizer.fit_on_texts(text_sequences)
sequences = tokenizer.texts_to_sequences(text_sequences)


```

```python
vocabulary_size = len(tokenizer.word_counts)
sequences = np.array(sequences)
```

### Creating an LSTM based model

```python
import keras
from keras.models import Sequential
from keras.layers import Dense,LSTM,Embedding
```

```python
# create your own custom model
def create_model(vocabulary_size, seq_len):
    
    model = Sequential()
    model.add(Embedding(vocabulary_size, 25, input_length=seq_len))
    #this has to always be the firstb layer
    #we are defining an input dimension for the embedding which is the entirety of the vocabulary and also the output dimension which is just the sequence length.
    #all what embedding is doing is to turn postive integers or indexes into dense vector of a fixed size(which is similar to normalizing or reducing the values into an integer below 1)
    #eg. [[4], [20]] -> [[0.6, -0.2]]
    model.add(LSTM(150, return_sequences=True))
    #we are using 150 neurons but will be slower compare to 50 neurons
    model.add(LSTM(150)) #another layer
    model.add(Dense(150, activation='relu')) #dense layer
    #we want to translate into vocabulary words because is just going to split out a number
    #so we want a dense network of our vocabulary size where the activation is soft max 
    model.add(Dense(vocabulary_size, activation='softmax'))
    #categorical_crossentropy means we are treating each vocabulary word as it's own individual category
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
   
    model.summary()
    
    return model
```

```python
from keras.utils import to_categorical
```

```python
# Train / Test Split
X = sequences[:,:-1]
y = sequences[:,-1]
y = to_categorical(y, num_classes=vocabulary_size+1)
seq_len = X.shape[1]
```

```python
from pickle import dump,load
```

```python
# define model
model = create_model(vocabulary_size+1, seq_len)
# fit model
model.fit(X, y, batch_size=128, epochs=30,verbose=1)
```

```python
# save the model to file
model.save('epoch30.h5')
# save the tokenizer
dump(tokenizer, open('epoch30', 'wb'))
```

### Generating New Text

```python
from random import randint
from pickle import load
from keras.models import load_model
from keras.preprocessing.sequence import pad_sequences
```

```python
def generate_text(model, tokenizer, seq_len, seed_text, num_gen_words):
    '''
    INPUTS:
    model : model that was trained on text data
    tokenizer : tokenizer that was fit on text data
    seq_len : length of training sequence
    seed_text : raw string text to serve as the seed
    num_gen_words : number of words to be generated by model
    '''
    
    # Final Output
    output_text = []
    
    # Intial Seed Sequence
    input_text = seed_text
    
    # Create num_gen_words
    for i in range(num_gen_words):
        
        # Take the input text string and encode it to a sequence
        encoded_text = tokenizer.texts_to_sequences([input_text])[0]
        
        # Pad sequences to our trained rate (50 words in the video)
        pad_encoded = pad_sequences([encoded_text], maxlen=seq_len, truncating='pre')
        
        # Predict Class Probabilities for each word
        pred_word_ind = model.predict_classes(pad_encoded, verbose=0)[0]
        
        # Grab word
        pred_word = tokenizer.index_word[pred_word_ind] 
        
        # Update the sequence of input text (shifting one over with the new word)
        input_text += ' ' + pred_word
        
        output_text.append(pred_word)
        
    # Make it look like a sentence.
    return ' '.join(output_text)
```
