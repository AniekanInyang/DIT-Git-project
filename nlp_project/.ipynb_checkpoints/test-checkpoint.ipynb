```python
import spacy
nlp = spacy.load('en_core_web_sm')

with open('google-phd-program.txt', encoding="utf8") as f:
    doc = nlp(f.read())
doc
```

```python
from spacy.matcher import Matcher
matcher = Matcher(nlp.vocab)
```

```python
pattern1 = [{'LOWER': 'ai'}]
# pattern2 = [{'LOWER': 'Artificial Intelligence'}]
pattern2 = [{'LOWER': 'artificial'}, {'LOWER': 'intelligence'}]
pattern3 = [{'LOWER': 'Artificial'}, {'IS_SPACE': True, 'OP':'*'},  {'LOWER': ' Intelligence'}]
# pattern5 = [{'LOWER': 'Artificial Intelligence'}, {'LOWER': 'ai'}]
# pattern3 = [{'LOWER': 'swimming'}, {'LOWER': 'vigorously'}]

matcher.add('artificial', None, pattern1, pattern2, pattern3)
```

```python
found_matches = matcher(doc)
print(found_matches)
```

```python
import pandas as pd
art = pd.read_csv("articles.csv")
# art.head(5)
blanks = []  # start with an empty list
# will return a tuples of index label review text
for i,lb,rv in art.itertuples():  # iterate over the DataFrame
    if type(rv)==str:            # avoid NaN values
        if rv.isspace():         # test 'review' for whitespace
            blanks.append(i)     # add matching index numbers to the list
        
# print(len(blanks), 'blanks: ', blanks)
art.drop(blanks, inplace=True)
data = art["articles"]
new_art = data.value_counts()
new_art
```

```python
import re
regex = r'([\w]+@[\w]+)'
def getEmails(new_art):
    for email in regex:
        if email:
            return re.findall(regex, new_art, re.M|re.I)
        else:
            print("no emails")
#         return re.findall(regex, new_art, re.M|re.I)
print(getEmails(new_art.to_string()))
```
